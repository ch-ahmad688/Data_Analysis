### Total cities

### Total population

```python
df = pd.read_csv('Pakistan_Population_100_Rows.csv')
print(df['Population'].sum())
```

### Average population

```python
df = pd.read_csv('Pakistan_Population_100_Rows.csv')
print(df['Population'].mean())
```

### Highest populated city

```python
df = pd.read_csv('Pakistan_Population_100_Rows.csv')
population = df['Population'].max()
print(df[df['Population'] == population][['City', 'Population']])
```

### Lowest populated city

```python
df = pd.read_csv('Pakistan_Population_100_Rows.csv')
population = df['Population'].min()
print(df[df['Population'] == population][['City', 'Population']])
```

### Province-wise total population

```python
df = pd.read_csv('Pakistan_Population_100_Rows.csv')
province_wise = df.groupby('Province')['Population'].sum().sort_values(ascending=False)
print(province_wise)
```

### Province-wise average literacy rate

```python
df = pd.read_csv('Pakistan_Population_100_Rows.csv')
province_wise = df.groupby('Province')['Literacy_Rate'].mean().sort_values(ascending=False)
print(province_wise)
```

### Top 10 populated cities

### Top 5 fastest growing cities
