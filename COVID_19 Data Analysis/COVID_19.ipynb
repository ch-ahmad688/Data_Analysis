```python
import pandas as pd
```

### Total confirmed cases

```python
df = pd.read_csv("covid_data_100_rows.csv")
print(df["Confirmed"].sum())
```

### Highest death rate Countries

```python
df = pd.read_csv("covid_data_100_rows.csv")
print(df.groupby('Country')['Deaths'].sum().sort_values(ascending=False).head(5))
```

### Recovery rate calculate

```python
df = pd.read_csv("covid_data_100_rows.csv")
df["Recovery Rate"] = (df['Recovered']/df["Confirmed"])*100
print(df[['Country','Confirmed','Recovered','Recovery Rate']])
```

### Country with most active cases

```python
df = pd.read_csv("covid_data_100_rows.csv")
print(df.groupby('Country')['Active'].sum().sort_values(ascending=False).head(5))
```

### Average daily confirmed cases

```python
df = pd.read_csv("covid_data_100_rows.csv")
print('Average daily cases')
print(df.groupby('Date')['Confirmed'].sum().mean())
```

### Pakistan Total Data

```python
df = pd.read_csv("covid_data_100_rows.csv")
print('Pakistan overall data')
print(df[df['Country'] == 'Pakistan'].describe())
```

### Top 5 countries by confirmed cases

```python
df = pd.read_csv("covid_data_100_rows.csv")
print(df.groupby('Country')['Confirmed'].sum().sort_values(ascending=False).head(5))
```

### Date-wise total confirmed cases

```python
df = pd.read_csv("covid_data_100_rows.csv")
print(df.groupby('Date')['Confirmed'].sum())
```

### Vaccination-to-Confirmed Ratio

```python
df = pd.read_csv("covid_data_100_rows.csv")
df['Vacnation Ratio'] = df['Vaccinated']/df['Confirmed']
print(df[['Country','Confirmed','Vaccinated','Vacnation Ratio']])
```

### Difference between Vaccinated and Confirmed

```python
df = pd.read_csv("covid_data_100_rows.csv")
df['Difference'] = df['Vaccinated']-df['Confirmed']
print(df[['Country','Confirmed','Vaccinated','Difference']])
```

### Missing values check

```python
df = pd.read_csv("covid_data_100_rows.csv")
print("Colum wise missing valuse")
print(df.isnull().sum())
print("Total Missing valuse")
print(df.isnull().sum().sum())
```

### Duplicated values check

```python
df = pd.read_csv("covid_data_100_rows.csv")
print(df.duplicated().sum())
```
