## Numpy array and basics

```python
import numpy as np
```

### create array from list

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:", arr_2d)
```

### list vs numpy array

```python
py_list = [1, 2, 3,]
print("Python list multiplication:", py_list*2)

py_array = np.array([1, 2, 3])
print("Python array multiplication:", py_array*2)

import time
start= time.time()
py_list = [i*2 for i in range(1000000)]
print("Time taken by Python list:", time.time() - start)

start = time.time()
py_array = np.arange(1000000)*2
print("Time taken by NumPy array:", time.time() - start)
```

### creating array from scratch

```python
zeros=np.zeros((3,4))
print ("zeros array :\n", zeros)

ones=np.ones((2,3))
print ("ones array :\n", ones)

full=np.full((2,2), 7)
print ("full array :\n", full)

random_array = np.random.rand(2,3)
print("Random array:\n", random_array)

sequence_array = np.arange(0, 11, 2)
print("Sequence array:\n", sequence_array)
```

### Vector,Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector:", vector)

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", matrix)

tensor = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("Tensor:\n", tensor)
```

### Array properties

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
print("Size:", arr.size)
print("Data type:", arr.dtype)
```

### Array Reshaping

```python
arr = np.arange(12)
print("Original array:", arr)

reshaped_arr = arr.reshape((3, 4))
print("Reshaped array:\n", reshaped_arr)

#flattened (returns a copy of the array collapsed into one dimension)
flattened_arr = reshaped_arr.flatten()
print("\n Flattened array:", flattened_arr)

#ravel (returns a flattened view of the array)
raveled_arr = reshaped_arr.ravel()
print("\n Raveled array:", raveled_arr)

transposed_arr = reshaped_arr.T
print("\n Transposed array:\n", transposed_arr)
```
