```python
import pandas as pd

df = pd.read_csv("delivered.csv")

runs = df.groupby("striker")["batsman_runs"].sum().sort_values(ascending=False)

print(runs.head(10))

plt.figure(figsize=(10,5))
plt.bar(runs.index, runs.values)
plt.title("Top 10 Run Scorers")
plt.xlabel("Batsman")
plt.ylabel("Total Runs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("delivered.csv")

wickets = df.groupby("bowler")["is_wicket"].sum().sort_values(ascending=False)

print(wickets.head(10))

plt.figure(figsize=(10,5))
plt.bar(wickets.index, wickets.values)
plt.title("Top 10 Wicket Takers")
plt.xlabel("Bowler")
plt.ylabel("Wickets")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sixes = (
    df[df["batsman_runs"] == 6]
    .groupby("striker")["batsman_runs"]
    .count()
    .sort_values(ascending=False)
)

print(sixes.head(10))

plt.figure(figsize=(10,5))
plt.bar(sixes.index, sixes.values)
plt.title("Top 10 Six Hitters")
plt.xlabel("Batsman")
plt.ylabel("Number of Sixes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fours = (
    df[df["batsman_runs"] == 4]
    .groupby("striker")["batsman_runs"]
    .count()
    .sort_values(ascending=False)
)

print(fours.head(10))

plt.figure(figsize=(10,5))
plt.bar(fours.index, fours.values)
plt.title("Top 10 Four Hitters")
plt.xlabel("Batsman")
plt.ylabel("Number of Fours")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
