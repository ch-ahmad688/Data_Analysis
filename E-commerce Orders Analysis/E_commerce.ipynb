### Total orders

```python
df = pd .read_csv('ecommerce_orders_100.csv')
total_order = len(df)
print(total_order)
```

### Best Selling Product

```python
product_sales = df.groupby("Product")["Quantity"].sum()

print("Best Selling Product:", product_sales.idxmax())
print("Total Quantity Sold:", product_sales.max())
```

### Profitable category

```python
df["Profit"] = (df["Unit_Price"]/df["Discount"]) * 100

profit_by_category = (
    df.groupby("Category")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

print(profit_by_category)
```

### City Wise sales

```python
df["Sales"] = (df["Quantity"]*df["Unit_Price"])-df["Discount"]
total_sales = df["Sales"].sum()
print("Total Sales:")
print(total_sales)
City_Wise_sales = df.groupby("City")["Sales"].sum()
print(City_Wise_sales)
```

### Payment method

```python
count_payment_method = df['Payment_Method'].value_counts()
print(count_payment_method)
```

### Monthly sales trend

```python
monthly_sales  = df['Order_Date'].value_counts()
print(monthly_sales)
```

### Top customers

```python
top_customers= df['Customer_Name'].value_counts()
print(top_customers)
```

### Average order value

```python
average_amount= df['Total_Amount'].mean()
print(average_amount)
```

### Cancelled Percentage

```python
cancelled_percentage = (
    df["Order_Status"]
    .value_counts(normalize=True)["Cancelled"] * 100
)

print(f"Cancelled Percentage: {cancelled_percentage:.2f}%")
```
