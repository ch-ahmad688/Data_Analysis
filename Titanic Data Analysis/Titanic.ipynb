### Female Survival Rate

```python
female = df[df["Sex"] == "female"]

female_survived = female["Survived"].sum()
female_ratio = (female_survived / len(female)) * 100

print("Female Survival Rate:", round(female_ratio, 2), "%")
```

### Male Survival rate

```python
male = df[df["Sex"] == "male"]

male_survived = male["Survived"].sum()
male_ratio = (male_survived / len(male)) * 100

print("Male Survival Rate:", round(male_ratio, 2), "%")
```

### Survive ratio with Pclass wise
