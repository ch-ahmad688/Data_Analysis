### Total Quality Sold

```python
df = pd.read_csv("sales.csv")
total_sold = df['Quality'].sum()
print("Total Quality Sold:")
print(total_sold)
```

### Total Order

```python
df = pd.read_csv("sales.csv")
total_order = df['Order_ID'].max()
print("Total Order:")
print(total_order)
```

### Total Sales

```python
df = pd.read_csv("sales.csv")
df["Sales"] = (df["Quality"]*df["Unit_Price"])-df["Discount"]
total_sales = df["Sales"].sum()
print("Total Sales:")
print(total_sales)
```

### Category Wise Sale

```python
Category_Wise_sales = df.groupby("Category")["Sales"].sum()
print(Category_Wise_sales)
```

### Product Wise Sale

```python
Product_Wise_sales = df.groupby("Product")["Sales"].sum()
print(Product_Wise_sales)
```

### City Wise Sale

```python
City_Wise_sales = df.groupby("City")["Sales"].sum()
print(City_Wise_sales)
```

### Payment Method Analysis

```python
payment_count = df.groupby("Payment_Method")["Order_ID"].count().sort_values(ascending=True)
print('Payment Count by Method (Descending):')
print(payment_count)
```

### Highest Spending Customer

```python
highest_Spender = df.groupby("Customer_Name")["Sales"].sum().sort_values(ascending=False)
print('Highest Spendin Customer:')
print(highest_Spender.head())
```

### Average Order Value

```python
Average_Order= total_sales/df["Order_ID"]
print("Average Order value :")
print(Average_Order.mean())
```

### Discount Analysis

```python
Discount_Analysis= df['Discount'].describe()
print(Discount_Analysis)
```

### Most Popular payment Method

```python
popular_payment = df['Payment_Method'].value_counts()
print('Payment Count by Method (Descending):')
print(popular_payment.head(1))
```

### Top Salesperson

```python
highest_Saleperson = df.groupby("Salesperson")["Sales"].sum().sort_values(ascending=False)
print('Highest Spendin Customer:')
print(highest_Saleperson.head(2))
```
