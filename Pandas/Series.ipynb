### Aggregation

### Grouped Method

```python
data = {
    "Name": ['Arun', 'Bala', 'Chitra','Ahmad','Ali'],
    "Age": [28, 34, 22, 22,34],
    "Salary":[10000, 20000, 30000, 40000, 50000]
}

df = pd.DataFrame(data)

grouped = df.groupby('Age')['Salary'].sum()
print(grouped)
```

### Merging

```python
# customer dataframe
df_customer = pd.DataFrame({
    'CustomerID': [1, 2, 3, ],
    'Name': ['Alice', 'Bob', 'Charlie'],
})
# order dataframe
df_orders = pd.DataFrame({
    'CustomerID': [1, 2, 4],
    'OrderAmount': [100, 200, 300]
})
# merge
df_merged = pd.merge(df_customer, df_orders, on='CustomerID', how='inner')
print('Inner Join:')
print(df_merged)
```

```python
# customer dataframe
df_customer = pd.DataFrame({
    'CustomerID': [1, 2, 3, ],
    'Name': ['Alice', 'Bob', 'Charlie'],
})
# order dataframe
df_orders = pd.DataFrame({
    'CustomerID': [1, 2, 4],
    'OrderAmount': [100, 200, 300]
})
# merge
df_merged = pd.merge(df_customer, df_orders, on='CustomerID', how='outer')
print('Outer Join:')
print(df_merged)
```

```python
# customer dataframe
df_customer = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
})
# order dataframe
df_orders = pd.DataFrame({
    'CustomerID': [1, 2, 4],
    'OrderAmount': [100, 200, 300]
})
# merge - cross join (Cartesian product)
df_merged = pd.merge(df_customer, df_orders, how='cross')
print('Cross Join:')
print(df_merged)
```

### Concatenation

```python
# region 1
df_Region1 = pd.DataFrame({
    'CustomerID': [1, 2],
    'Name': ['Alice', 'Bob'],
})

# region 2
df_Region2 = pd.DataFrame({
    'CustomerID': [3, 4],
    'Name': ['Charlie', 'David'],
})

# concatenate vertically
df_concat = pd.concat([df_Region1,df_Region2],ignore_index=True)
print(df_concat)
```

```python
# region 1
df_Region1 = pd.DataFrame({
    'CustomerID': [1, 2],
    'Name': ['Alice', 'Bob'],
})

# region 2
df_Region2 = pd.DataFrame({
    'CustomerID': [3, 4],
    'Name': ['Charlie', 'David'],
})

# concatenate horizontally
df_concat = pd.concat([df_Region1,df_Region2],axis=1,ignore_index=True)
print(df_concat)
```
